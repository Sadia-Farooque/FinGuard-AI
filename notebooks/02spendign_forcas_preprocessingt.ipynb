```python
# ================================
# IMPORTS
# ================================
import pandas as pd
import numpy as np



import os
import pickle

print("all done")


# ================================
# LOAD DATA
# ================================
df = pd.read_csv("../data/retail_sales_dataset.csv")
df.head()


# ================================
# BASIC CLEANING
# ================================
df = df.drop_duplicates()
df.isnull().sum()


# ================================
# DATE PARSING + FEATURE ENGINEERING
# ================================
# assuming column name = 'Date'
df["Date"] = pd.to_datetime(df["Date"])

# extracting time-based features
df["day_of_week"] = df["Date"].dt.dayofweek
df["month"] = df["Date"].dt.month


# ================================
# ROLLING FEATURE (7-day avg)
# ================================
df = df.sort_values("Date")

df["rolling_7day_avg"] = df["Total Amount"].rolling(window=7).mean()

# handle NaNs created by rolling
df["rolling_7day_avg"].bfill(inplace=True)



# ================================
# ENCODING CATEGORICAL FEATURES
# ================================
# Pure pandas Label Encoding
cols_to_encode = ['Gender', 'Product Category'] 

for col in cols_to_encode:
    df[col] = df[col].astype("category").cat.codes



# ================================
# FEATURE SCALING (OPTIONAL but aligned)
# Pure pandas scaling
cols_to_scale = ['Age', 'Quantity', 'Price per Unit'] 

for col in cols_to_scale:
    df[col] = (df[col] - df[col].mean()) / df[col].std()



# ================================
# SPLITTING DATA
# ================================
# Pure pandas train/test split
train_df = df.sample(frac=0.8, random_state=42) # 80% for training
test_df = df.drop(train_df.index)               # 20% for testing

# Separate X and Y
X_train = train_df.drop("Total Amount", axis=1) # Replace with your target column
y_train = train_df["Total Amount"]

X_test = test_df.drop("Total Amount", axis=1)
y_test = test_df["Total Amount"]



# ================================
# CREATE PROCESSED FOLDER
# ================================
os.makedirs("../data/processed", exist_ok=True)


# ================================
# SAVE DATA (PICKLE)
# ================================
with open("../data/processed/spend_X_train.pkl", "wb") as f:
    pickle.dump(X_train, f)

with open("../data/processed/spend_X_test.pkl", "wb") as f:
    pickle.dump(X_test, f)

with open("../data/processed/spend_y_train.pkl", "wb") as f:
    pickle.dump(y_train, f)

with open("../data/processed/spend_y_test.pkl", "wb") as f:
    pickle.dump(y_test, f)


```
